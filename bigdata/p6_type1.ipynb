1. 주어진 데이터는 각 소방서의 출동/도착 시간데이터이다.
출동시간과 도착시간 차이가 평균적으로 가장 오래 걸린 소방서의 시간을 분으로 변환해 출력하시오. (반올림 후 정수 출력)

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_1/data6-1-1.csv")
df.head()
```

```python
# df.info()
df['출동시간'] = pd.to_datetime(df['출동시간'])
df['도착시간'] = pd.to_datetime(df['도착시간'])
df.info()
```

```python
df['차이'] = df['도착시간'] - df['출동시간']
sec = df.groupby('소방서')['차이'].mean().nlargest(1).values[0] / (10 ** 9)
round(int(sec) / 60)
# 81분
```

2. 학교에서 교사 한 명당 맡은 학생 수가 가장 많은 학교를 찾고, 그 학교의 전체 교사 수를 구하시오. (정수 출력)

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_1/data6-1-2.csv")
df.head()
```

```python
# df['비율'] = (df['1학년'] + df['2학년'] + df['3학년'] + df['4학년'] + df['5학년'] + df['6학년']) / df['교사수'] # 코드 축약 가능?
# df['비율'] = df.loc[:, '1학년':'6학년'].sum(axis=1) / df['교사수']
df['비율'] = df.iloc[:, 2:].sum(axis=1) / df['교사수']
df.head()
df.nlargest(1, '비율')['교사수']
# 19
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_1/data6-1-3.csv")
df.head()
```

```python
# df['합계'] = df['강력범죄'] + df['절도범죄'] + df['폭력범죄'] + df['지능범죄'] + df['풍속범죄'] + df['교통범죄'] # 코드 축약 가능?
# df['합계'] = df.loc[:, '강력범죄':'교통범죄'].sum(axis=1)
df['합계'] = df.iloc[:, 1:7].sum(axis=1)
df['연도'] = df['날짜'].str[0:4]
df.groupby('연도')['합계'].mean().nlargest(1)
# df.nlargest(1, '합계')
# df.head()
cond = df['연도'] == "2023"
round(df[cond]['합계'].mean())
# 531
```
