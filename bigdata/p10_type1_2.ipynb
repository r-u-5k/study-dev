### 문제 1
제공된 데이터(subject_performance.csv)는 학생들이 각 문제를 풀었을 때의 정답 여부를 기록한 자료이다. 각 문제는 대주제와 소주제로 분류되어 있다. 아래 수행 순서에 따라 데이터를 처리한 뒤, 구한 값을 [제출 형식]에 맞춰 답안 페이지에 입력하시오.

제공 데이터
- major_topic : 대주제
- sub_topic : 소주제
- student_id : 학생 ID
- is_correct : 정답 여부(1: 정답, 0: 오답)

수행 순서
1. 소주제별 정답률을 계산한다.
    - 정답률 = (해당 소주제의 정답 개수) / (해당 소주제의 전체 문제 풀이 수)
2. 소주제들의 정답률 중 3번째로 높은 값을 구하시오.
    - 단, 정답률이 같은 경우 하나로 취급한다. 정답률이 0.7, 0.7, 0.6, 0.5인 경우, 세 번째로 높은 값은 0.5이다.

[제출형식] 반올림하여 소수 셋째 자리까지 작성


```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/10_1/subject_performance.csv")
df.head()
```

```python
df.groupby(['major_topic', 'sub_topic'])['is_correct'].mean().nlargest(4)
round(0.673913, 3)
# 0.674
```

### 문제2
제공된 데이터(cafe_sales.csv)는 2024년 4월부터 2025년 12월까지의 카페 메뉴 판매 이력이다. 소문항을 각각 구한 다음 [제출 형식]에 맞춰 답안 페이지에 입력하시오.

제공 데이터
- order_date : 주문 날짜 (예 2024.05.24 13:22)
- category : 음료/디저트 종류 (예 coffee, tea …)
- item : 메뉴명
- price : 판매 금액

소문항

2-1. 연-월별 총 매출액을 계산하여 큰 순서대로 정렬했을 때, 2번째로 큰 매출액을 입력하시오.

2-2. 연-월별 총 매출액이 4번째로 큰 연-월을 찾으시오. 해당 연-월의 데이터에서 카테고리별 매출 합계를 계산하고, 그중 최댓값을 입력하시오.

[제출 형식] 정수로 작성


```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/10_1/cafe_sales.csv")
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df.head()
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df.head()
```

```python
df.groupby(['year', 'month'])['price'].sum().sort_values(ascending=False)
# 328741
# 2024년 10월
```

```python
cond = (df['year'] == 2024) & (df['month'] == 10)
df2 = df[cond]
df2.head()
```

```python
df2.groupby('category')['price'].sum().max()
# 104660
```

### 문제3
제공된 데이터(hamspam.csv)는 문자 메시지 내용과 스팸 여부를 담고 있다.
아래 수행 순서에 따라 데이터를 처리한 뒤, 구한 값을 [제출 형식]에 맞춰 답안 페이지에 입력하시오.

제공 데이터
- label : 메시지 구분 (spam: 스팸, ham: 정상)
- text : 메시지

수행 순서
1. 각 메시지의 단어 개수를 띄어쓰기(공백) 기준으로 계산하시오.
2. 스팸 메시지의 평균 단어 개수와 정상 메시지의 평균 단어 개수를 각각 구하시오.
3. 두 평균의 차이의 절댓값을 구하여 입력하시오.

[제출형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/10_1/hamspam.csv")
df.head()
```

```python
df['words'] = df['text'].str.split(' ')
df['count'] = df['words'].str.len()
df.head()
df.groupby('label')['count'].mean()
round(abs(10.070026 - 10.022919), 3)
# 0.047
```
